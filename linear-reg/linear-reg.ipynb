```python
import numpy as np
import seaborn as sns
```

```python
a = np.random.randint(10)
b = np.random.randint(10)
a, b
```

```python
m = 100
```

```python
X = 2 * np.random.rand(m, 1)
X
```

```python
y = a + b * X + np.random.randn(m, 1)
y
```

```python
X.reshape(1, -1)
```

```python
sns.scatterplot(X.reshape(1, -1)[0], y.reshape(1, -1)[0])
```

```python
bs = np.random.rand(2, 1)
bs
```

```python
X_b = np.c_[np.ones((m, 1)), X]
X_b
```

```python
eta = .3
for i in range(100):
  y_c = X_b.dot(bs)
  gradients = X_b.T.dot(y_c - y) / m
  bs = bs - eta * gradients
  print(bs)
```
